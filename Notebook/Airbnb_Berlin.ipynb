# **Checking for missing values**

```python
missing = df.isna().mean()*100
missing
```

```python
df["Host Response Time"] = df["Host Response Time"].fillna("No response")
```

```python
under_13 = missing[missing < 13].index
```

```python

df = df.dropna(subset= under_13)
```

```python
df.isna().mean()*100
```

```python
df = df.drop(columns=["Square Feet"])
```

# **Data Cleaning**

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.columns
```

```python
df = df.drop(columns=["review_id","reviewer_id","reviewer_name","listing_url","host_url"])
```

```python
df.columns
```

```python
numeric_cols = df.select_dtypes(include=["number"]).columns
numeric_cols
```

```python
df = df.drop(columns=["latitude", "longitude"])
```

NOTE: consider removing "listing_id" and "host_id"

```python
df["country"].unique()
```

```python
df['neighbourhood'].unique()
```

```python
df["neighbourhood"] = (
    df["neighbourhood"]
    .astype(str)
    .str.encode("latin1")
    .str.decode("utf-8")
)
```

```python
df['neighbourhood'].unique()
```

```python
text_cols = df.select_dtypes(include="object").columns

# Fix encoding for all
for col in text_cols:
    df[col] = (
        df[col].astype(str).str.encode("latin1", errors="ignore").str.decode("utf-8", errors="ignore"))
```

```python
df[['neighbourhood', 'neighborhood_group', 'city',
       'postal_code', 'country_code']]
```

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

# **Checking for Outliers**

```python
numeric_cols = df.select_dtypes(include="number").columns

# filter out identifiers
numeric_cols = numeric_cols.drop(["listing_id", "host_id"], errors="ignore")

numeric_cols
```

```python
for col in numeric_cols:
    plt.figure()
    df.boxplot(column=col)
    plt.title(col)
    plt.show()
```

```python
upper = df["price"].quantile(1)

plt.figure(figsize=(8,5))

sns.histplot(df[df["price"] <= upper]["price"], bins=30)

plt.title("Price Distribution (Up to 100th Percentile)")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

# **Analytics**

Do larger places charge more?

```python
sns.regplot(x="accomodates", y="price", data=df)
plt.title("Price vs Accommodates")
plt.show()
```

Relationship between prices and ratings

```python
sns.scatterplot(x="price", y="overall_rating", data=df)
plt.show()
```

Location

```python
plt.figure(figsize=(10,6))

sns.countplot(
    y="neighborhood_group",
    data=df,
    order=df["neighborhood_group"].value_counts().index
)

plt.title("Number of Listings per Neighborhood")
plt.xlabel("Count")
plt.ylabel("Neighborhood")

plt.tight_layout()
plt.show()
```
